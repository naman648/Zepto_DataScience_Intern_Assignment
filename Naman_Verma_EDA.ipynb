## Step-01: Loading the Dataset

```python
import pandas as pd
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

## Step-02: Overview of the Dataset

```python
print(customers.head())
print(products.head())
print(transactions.head())

# Checking for Missing/null values:
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())
```

### As you can see, there are no missing or null values on the Dataset, so we don't need to perform Data Cleaning process or Filling the null values.

## Step-03: Exploratory Data Analysis (EDA)

```python
# Summary
print(transactions.describe())
print(products.describe())

#Visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of transaction values
plt.figure(figsize=(10, 6))
sns.histplot(transactions['TotalValue'], bins=30, kde=True)
plt.title('Distribution of Transaction Values')
plt.xlabel('Total Value (USD)')
plt.ylabel('Frequency')
plt.show()

# Total sales by region
sales_by_region = transactions.groupby(customers['Region']).sum()['TotalValue']
sales_by_region.plot(kind='bar', title='Total Sales by Region')
plt.ylabel('Total Sales (USD)')
plt.show()
```

```python
# Selecting only numeric columns for correlation analysis
numeric_transactions = transactions.select_dtypes(include=['float64', 'int64'])
# Correlation-matrix
correlation = numeric_transactions.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
